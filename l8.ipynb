```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
#Task 1
#Create a line chart representing the trend of GDP per Capita over the years for a country of your choice. Ensure the line chart has a title and labelled axes.#
df = pd.read_csv('C:/Users/shoam/Downloads/matplotlib_charts_class_exc.csv')
country_data = df[df['Country'] == 'Canada']

plt.figure(figsize=(8, 5))
plt.plot(country_data['Year'], country_data['GDP per Capita'])

plt.title(f'GDP per Capita Trend in Canada')
plt.xlabel('Year')
plt.ylabel('GDP per Capita')

plt.grid(True)
plt.show()

```

```python
#Task 2
#Using a histogram, visualize the distribution of the GDP per Capita values for a particular year of your choice.#
year_data = df[df['Year'] == 2003]


plt.figure(figsize=(8, 5))
plt.hist(year_data['GDP per Capita'], bins=10)


plt.title(f'Distribution of GDP per Capita in 2003')
plt.xlabel('GDP per Capita')
plt.ylabel('Number of Countries')



plt.show()
```

```python
#Task 3
#Plot a bar chart to compare the GDP per Capita for five different countries in the year 2010.#
data_2010 = df[df['Year'] == 2010]

plt.figure(figsize=(8, 5))
plt.bar(data_2010['Country'], data_2010['GDP per Capita'])


plt.title('GDP per Capita Comparison in 2010')
plt.xlabel('Country')
plt.ylabel('GDP per Capita')


plt.xticks(rotation=45)


plt.show()
```

```python
#Task 4
#Using a scatter plot, explore the relationship, if any, between the year and GDP per Capita for a chosen country.
country_data = df[df['Country'] == 'USA']


plt.figure(figsize=(8, 5))
plt.scatter(country_data['Year'], country_data['GDP per Capita'])

plt.title(f'GDP per Capita vs Year for USA')
plt.xlabel('Year')
plt.ylabel('GDP per Capita')


plt.show()
```
